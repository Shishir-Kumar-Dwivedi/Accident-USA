```python
import pandas as pd

# importing csv file
df = pd.read_csv('acc_16.csv')

# Set display option to show more columns
pd.set_option('display.max_columns', None)  # None means unlimited columns

# Display the entire DataFrame
df.head()
```

```python
# List of column names to be deleted
columns_to_delete = ['PSU', 'PJ', 'STRATUM','VE_FORMS','PVH_INVL','PEDS','PERMVIT','PERNOTMVIT','MAN_COLL','RELJCT1','RELJCT2','TYP_INT',
                    'CF1','CF2','CF3','LGTCON_IM','PSUSTRAT','PSU_VAR']
df = df.drop(columns=columns_to_delete, axis=1)
df.head()
```

```python
column_mapping = {'VE_TOTAL': 'Total_Vehicle', 'NUM_INJ': 'Injured_person','HARM_EV':'HARMED_ElectricVehicle','ALCOHOL':'ACCED_DUE_TO_ALCOHOL'
                 ,'SCH_BUS':'SCHOOL_BUS_HARMED','URBANICITY':'URBAN_CITY','WKDY_IM':'WEEKENDDAYS_IM'}
df.rename(columns=column_mapping, inplace=True)
df.head()
```
